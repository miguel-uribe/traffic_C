```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import animation
import numpy as np
import time
%matplotlib notebook
```

```python
data = np.loadtxt('sim_results/sim_results_5_2_left_100m_50_0.20_0.50_0.20_0.40_0.10_0.30_cardata.txt')
```

```python
fig, ax = plt.subplots(figsize=(8,3))
grph, = ax.plot([],[], 'o')
ax.set_xlim(3500,4500)
ax.set_ylim(-0.5,5.5)
#timetext = ax.text(180,5, '%d'%0)

t0 = 6000
grph.set_data(data[data[:,0]==t0,1],data[data[:,0]==t0,2])


def animate(t):
    grph.set_data(data[data[:,0]==t+t0,1],data[data[:,0]==t+t0,2])
    #timetext.set_text('%d'%t)
    return [grph]
```

```python
# blit=True re-draws only the parts that have changed.
anim = animation.FuncAnimation(fig, animate, frames=20*60, interval=100, blit=True)
```
