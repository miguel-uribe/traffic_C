```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import animation
import numpy as np
import time
%matplotlib notebook
```

```python
data = np.loadtxt('cardata.txt')
```

```python
fig, ax = plt.subplots(figsize=(8,3))
grph, = ax.plot([],[], 'o')
ax.set_xlim(0,200)
ax.set_ylim(-0.5,5.5)
timetext = ax.text(180,5, '%d'%0)

def animate(t):
    grph.set_data(data[data[:,0]==t,1],data[data[:,0]==t,2])
    timetext.set_text('%d'%t)
    return [grph]
```

```python
# blit=True re-draws only the parts that have changed.
anim = animation.FuncAnimation(fig, animate, frames=200, interval=100, blit=True)
```

```python
from IPython.display import HTML
HTML(anim.to_html5_video())
```
